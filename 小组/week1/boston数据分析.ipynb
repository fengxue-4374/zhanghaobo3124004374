```python
# 导入pandas库的read_csv函数，用于读取CSV文件
from pandas import read_csv

# 定义文件路径和列名
filename = r'D:\\firstweek\\housing.csv'
# 使用read_csv函数读取CSV文件，指定文件路径、列名和分隔符
names=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
# 使用read_csv函数读取CSV文件，指定文件路径、列名和分隔符
dataset = read_csv(filename, names=names, sep=r'\s+')
```

```python
# 导入pandas库的set_option函数，用于设置显示选项
from pandas import set_option
# 打印数据集的形状，即行数和列数
print(dataset.shape)
# 打印数据集各列的数据类型
print(dataset.dtypes)
# 设置显示宽度，以便更好地查看数据集的前几行
set_option('display.width',120)
# 打印数据集的前30行，以查看数据概览
print(dataset.head(30))

# 设置显示精度为1位小数，减少输出的复杂性
set_option('display.precision',1)
# 打印数据集的描述性统计信息，包括计数、平均值、标准差、最小值和四分位数
print(dataset.describe())
# 设置显示精度为2位小数，适用于相关性矩阵的显示
set_option('display.precision',2)
# 打印数据集的相关性矩阵，使用皮尔逊相关系数，以分析列之间的线性关系
print(dataset.corr(method='pearson'))
```

```python
# 导入numpy和matplotlib的pyplot，以及pandas的散点图矩阵工具
import numpy as np
from numpy import arange
from matplotlib import pyplot
from pandas.plotting import scatter_matrix

# 为数据集的每一列绘制直方图，以观察数据分布
dataset.hist(sharex=False,sharey=False,xlabelsize=1,ylabelsize=1)
pyplot.show()
# 为数据集的每一列绘制密度图，以更精确地观察数据分布的形状
dataset.plot(kind='density',subplots=True,layout=(4,4),sharex=False,fontsize=1)
pyplot.show()
# 为数据集的每一列绘制箱线图，以观察数据的五数概括（最小值、下四分位数、中位数、上四分位数、最大值）
dataset.plot(kind='box',subplots=True,layout=(4,4),sharex=False,sharey=False,fontsize=8)
pyplot.show()
# 绘制散点图矩阵，以直观地观察每两列数据之间的关系
scatter_matrix(dataset)
pyplot.show()
```
